```python
from datetime import time as dt_time
from IPython.display import display
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import time
import random

data = pd.read_csv("wine.csv")

def beb_pd(data):
    # Створюємо стовпчик "якість" з випадковими значеннями "низька" та "висока"
    quality_values = ['low', 'hight']
    data['quality'] = [random.choice(quality_values) for _ in range(len(data))]

    
    return data
data = beb_pd(data)

display(data)

```

<h2>Заваньтаження даних та додавання нової колонки<h2>

```python

import numpy as np


df = np.genfromtxt('wine.csv', delimiter=',', skip_header=1)

def beb_np(df):
    # Створюємо стовпчик "quality" з випадковими значеннями "low" та "high"
    quality_values = np.random.choice(['low', 'high'], size=len(df))
    
    # Зберігаємо як масив рядків (object)
    quality_values = quality_values.astype(object)
    
    # Об'єднання даних
    data_with_quality = np.column_stack((df, quality_values))
    
    return data_with_quality

df = beb_np(df)

display(df)

```

<h2>Пронормувати вибраний датасет або стандартизувати його
(нормалізація і стандартизація мають бути реалізовані як окремі 4 функції без застосування додаткових бібліотек, як наприклад sklearn.preprocessing).<h2>

```python
def norm_pd(data):
    min_acidity = data['fixed_acidity'].min()
    max_acidity = data['fixed_acidity'].max()
    mean_acidity = data['fixed_acidity'].mean()
    std_acidity = data['fixed_acidity'].std()
    
    data['volatile_acidity'] = (data['fixed_acidity'] - min_acidity) / (max_acidity - min_acidity)
    data['citric_acid'] = (data['fixed_acidity'] - mean_acidity) / std_acidity

    return data


def norm_np(df):
    # Отримуємо мінімальне та максимальне значення для fixed_acidity
    min_acidity = np.min(df[:, 0])
    max_acidity = np.max(df[:, 0])
    
    # Отримуємо середнє значення та стандартне відхилення для fixed_acidity
    mean_acidity = np.mean(df[:, 0])
    std_acidity = np.std(df[:, 0])
    
    # Нормалізація volatile_acidity
    df[:, 1] = (df[:, 0] - min_acidity) / (max_acidity - min_acidity)
    
    # Нормалізація citric_acid
    df[:, 2] = (df[:, 0] - mean_acidity) / std_acidity
    
    return df



start_time = time.time()
display(norm_pd(data)[['fixed_acidity','volatile_acidity','citric_acid']])
end_time_pd = time.time() - start_time
start_time = time.time()
display(norm_np(df)[:, [0, 1, 2]])
end_time_np = time.time() - start_time
print('Pandas:',end_time_pd)
print('Nampy:',end_time_np)


```

<h2>5. Підрахувати коефіцієнт Пірсона та Спірмена для двох integer/real
атрибутів.<h2>

```python
from scipy.stats import pearsonr, spearmanr
def pir_pd(data):
    x = data['total_sulfur_dioxide']
    y = data['free_sulfur_dioxide']

  
    pearson_corr, pearson_p_value = pearsonr(x, y)


    spearman_corr, spearman_p_value = spearmanr(x, y)

    print("Коефіцієнт Пірсона:", pearson_corr)
    print("Коефіцієнт Спірмена:", spearman_corr)

def pir_np(df):
    x = df[:, 6]
    y = df[:, 5]

    pearson_corr, pearson_p_value = pearsonr(x, y)

    
    spearman_corr, spearman_p_value = spearmanr(x, y)

    print("Коефіцієнт кореляції Пірсона:", pearson_corr)
    print("Коефіцієнт кореляції Спірмена:", spearman_corr)

start_time = time.time()
pir_pd(data)
end_time_pd = time.time() - start_time
start_time = time.time()
pir_np(df)
end_time_np = time.time() - start_time
print('Pandas:', end_time_pd)
print('Numpy:', end_time_np)
```

<h2>6. Провести One Hot Encoding категоріального string атрибуту.<h2>

```python
data_encoded = pd.get_dummies(data, columns=['quality'])

def hot_np(df):
    
    unique_values = np.unique(df[:, -1])

    
    one_hot_encoded = np.zeros((len(df), len(unique_values)), dtype=bool)

   
    for i, value in enumerate(unique_values):
        one_hot_encoded[:, i] = (df[:, -1] == value)

    
    df_encoded = np.hstack((df[:, :-1], one_hot_encoded))
    return df_encoded

start_time = time.time()
display(data_encoded)
end_time_pd = time.time() - start_time
start_time = time.time()
display(hot_np(df))
end_time_pd = time.time() - start_time
print('Pandas:',end_time_pd)
print('Nampy:',end_time_np)
```

<h2>7. Провести візуалізацію багатовимірних даних.<h2>

```python
import pandas as pd
import matplotlib.pyplot as plt

def correlation_heatmap(data):
    
    numeric_data = data.select_dtypes(include='number')
    
    plt.figure(figsize=(10, 6))
    corr = numeric_data.corr()
    plt.imshow(corr, cmap='coolwarm', interpolation='nearest')
    plt.colorbar()
    plt.title('Карта кореляції атрибутів вина')
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
    plt.yticks(range(len(corr.columns)), corr.columns)
    plt.show()
correlation_heatmap(data)
```
