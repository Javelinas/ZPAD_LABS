<h2>3. Обрати всі домогосподарства, у яких сила струму лежить в межах
19-20 А, для них виявити ті, у яких пральна машина та холодильних
споживають більше, ніж бойлер та кондиціонер.<h2>

```python

def hol_pd(data):
    holod = data[(data['Global_intensity'] >= 19) & (data['Global_intensity'] <= 20)]

    result = holod[(holod['Sub_metering_2'] > holod['Sub_metering_3'])]
  
    return result

def hol_np(df):
    holod = df[(df['Global_intensity'] >= 19) & (df['Global_intensity'] <= 20)]
    result_np = holod[(holod['Sub_metering_2'] > holod['Sub_metering_3'])]  
    return result_np

start_time = time.time()
hol_pd(data)
end_time_pd = time.time() - start_time
display(hol_pd(data))
start_time = time.time()
hol_np(df)
end_time_np = time.time() - start_time
display(hol_np(df))
print('Pandas:',end_time_pd)
print('Nampy:',end_time_np)
```

<h2>5. Обрати ті домогосподарства, які після 18-00 споживають понад 6
кВт за хвилину в середньому, серед відібраних визначити ті, у яких
основне споживання електроенергії у вказаний проміжок часу
припадає на пральну машину, сушарку, холодильник та освітлення
(група 2 є найбільшою), а потім обрати кожен третій результат із
першої половини та кожен четвертий результат із другої половини.<h2>

```python


def end_pd(data):
    end = data[(data['Time'].apply(lambda x: x > dt_time(hour=18))) & (data['Global_active_power'] > 6.0)]
    mor = end[(end['Sub_metering_2'] > end['Sub_metering_1']) &
                              (end['Sub_metering_2'] > end['Sub_metering_3'])]
   
    first_end = mor.head(len(mor) // 2)
    second_end= mor.tail(len(mor) // 2)
    result_df = pd.concat([first_end.iloc[::3, :], second_end.iloc[::4, :]])
    return result_df



import pandas as pd
import numpy as np
import datetime as dt

def end_np(df):
    
    target_time = dt.datetime.strptime('18:00:00', '%H:%M:%S').time()
    
    df_time = np.array([datetime.strptime(tm, '%H:%M:%S').time() for tm in df["Time"]])
    
    filtered_np = df[(df_time > target_time) & (df['Global_active_power'] > 6.0)]
    
    final_np = filtered_np[(filtered_np['Sub_metering_2'] > filtered_np['Sub_metering_1']) & (filtered_np['Sub_metering_2'] > filtered_np['Sub_metering_3'])]
   
    first_half = final_np[:len(final_np) // 2]
    second_half = final_np[len(final_np) // 2:]
    
    result_np = np.concatenate([first_half[::3], second_half[::4]])
    
    return result_np


start_time = time.time()
display(end_pd(data))
end_time_pd = time.time() - start_time
start_time = time.time()
display(end_np(df))
end_time_pd = time.time() - start_time
print('Pandas:',end_time_pd)
print('Nampy:',end_time_np)
```
